# PointPlot

### 사고심각도 가중치는 273/32/2/1 로 사용하였음

```python
len(old_pedestrian)
```

```python
fig, axes = plt.subplots(2,2)

sns.pointplot(x='요일',y='사고심각도', data=old_pedestrian, ax=axes[0][0])
sns.pointplot(x='주야',y='사고심각도', data=old_pedestrian, ax=axes[0][1])
sns.pointplot(x='사고내용',y='사고심각도', data=old_pedestrian, ax=axes[1][0])
sns.pointplot(x='사고유형',y='사고심각도', data=old_pedestrian, ax=axes[1][1])

fig.set_size_inches(15,10)
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(2,1,figsize=(25,20))

sns.pointplot(x='가해자성별',y='사고심각도', data=old_pedestrian, ax=ax1)
sns.pointplot(x='피해자성별',y='사고심각도', data=old_pedestrian, ax=ax2)

plt.tight_layout()
plt.show()
```
