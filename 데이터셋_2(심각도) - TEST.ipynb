# 분할표 만들어서 카이제곱검정

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

contingency = pd.crosstab(df['사고유형'],df['심각도'])
contingency
```

```python
chi, p, dof, expected = chi2_contingency(contingency)

print(f"chi 스퀘어 값: {chi}",
      f"p-value (0.05): {p}",
      f"자유도 수: {dof}",
      f"기대값: \n{pd.DataFrame(expected)}",
      f"측정값: \n{contingency}", sep = "\n" )
```

---

```python
contingency = pd.crosstab(df['연령가해자'],df['심각도'])
contingency
```

```python
chi, p, dof, expected = chi2_contingency(contingency)

print(f"chi 스퀘어 값: {chi}",
      f"p-value (0.05): {p}",
      f"자유도 수: {dof}",
      f"기대값: \n{pd.DataFrame(expected)}",
      f"측정값: \n{contingency}", sep = "\n" )
```

---

```python
contingency = pd.crosstab(df['차량용도가해자_중분류'],df['심각도'])
contingency
```

```python
chi, p, dof, expected = chi2_contingency(contingency)

print(f"chi 스퀘어 값: {chi}",
      f"p-value (0.05): {p}",
      f"자유도 수: {dof}",
      f"기대값: \n{pd.DataFrame(expected)}",
      f"측정값: \n{contingency}", sep = "\n" )
```

```python
chi2_check = []
categorical_columns=df.drop(['심각도'],axis=1).columns

for i in categorical_columns:
    if chi2_contingency(pd.crosstab(df['심각도'], df[i],normalize=False))[1] < 0.05:
        chi2_check.append('P<0.05이므로 연관있음')
    else:
        chi2_check.append('연관없음')
res = pd.DataFrame(data = [categorical_columns, chi2_check] ).T 
res.columns = ['Column', 'Hypothesis']
res
```
