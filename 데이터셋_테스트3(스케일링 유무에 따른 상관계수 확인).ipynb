## TEST2. 스케일링 후 

```python
result_1=result[['연령피해자','사고심각도']]
result_2 = result.drop(['연령피해자','사고심각도'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler 

minmax_scaler = StandardScaler()
result_1 = minmax_scaler.fit_transform(result_1)
result_1 = pd.DataFrame(result_1)
result_1 = result_1.rename({0:'연령피해자',1:'사고심각도'},axis='columns')
```

```python
result_set = pd.concat([result_1,result_2],axis=1)
result_set
```

```python
cor_data = result_set.corr()
```

```python
a = cor_data['사고심각도']
for i in range(len(a)):
    if a[i]>=0.2 and a[i]!=1:
        print(a[i])
```
